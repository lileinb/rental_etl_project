```python
# 
import pandas as pd
df = pd.read_csv(r"D:\Onediver\OneDrive\Desktop\rental_etl_project\Data\data_raw\houses.csv")
import matplotlib.pyplot as plt
import seaborn as sns
df['price_clean'] = (
	df['price']
	.str.replace('RM', '', regex=False)
	.str.replace(',', '', regex=False)
	.str.replace(' ', '', regex=False)
	.astype(float)
)

filtered = df[df['price_clean'] < df['price_clean'].quantile(0.95)]
plt.figure(figsize=(10, 6))
sns.histplot(filtered['price_clean'], bins=30, kde=True)
plt.title('Rent Price Distribution in malaysia')
plt.xlabel('Price(RM)')
plt.ylabel('Number of listings')
plt.grid(True)
plt.tight_layout()
plt.show(block=True)
```
